```python
# Задача 1: Создайте массив 10х10, заполненный в центре нулями, а по границе единицами.
import numpy as np
y=np.zeros((10,10))
y[0::9,::]=1
y[::,0::9]=1
print(y)
```

```python
# Задача 3: Создайте матрицу 5x5 со значениями 1,2,3,4 под главной диагональю. Остальные значения - нули
import numpy as np
A=np.zeros((5,5))
for i in range(5):
    A[i,i-1]=i
print(A)
```

```python
# Задача 2: Создайте массив 8х8, заполненный 0 и 1 в шахматном порядке.
x = np.zeros((8,8))
x[1::2,::2] = 1
x[::2,1::2] = 1
print(x)
```

```python
# Задача 4: Создайте случайный массив 5х5 и нормируйте его так, чтобы среднее по каждой строке было в точности ноль.
import numpy as np
x=np.random.random((5,5))
for y in x:
    xE = np.mean(y)
    xstd=np.std(y)
    y = (y - xE)/xstd
    print(y)
```

```python
#Задача 5: Создайте двумерный массив 10x3, заполненный случайными числами от 0 до 1. В каждой строке выберите значение, наиболее близкое к 0.5.
import numpy as np
y=np.random.uniform(0, 1, size=(10,3))
print(y)
for i in y:
    print(min(i,key=lambda x: abs(x-0.5)))
```

```python
# Задача 6: Создайте двумерный массив 6x6, заполненный случайными значениями. Сумму по каждой строке поделите на максимальный элемент из столбца с тем же индексом.
import numpy as np
y=np.random.random((6,6))
print(y)
print((np.sum(y,1)/np.max(y,0)))
```

```python
# Задача 7: На примере массива x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) найдите максимальный элемент в массиве среди элементов, перед которыми стоит ноль
import numpy as np
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
print(max((x[i] for i in range(1,len(x)) if x[i-1]==0), default=None))
```

```python
# Задача 8: Пусть заданы два массива x = np.ones(10) и i = np.array([0, 1, 2, 3, 5, 5, 5, 8]). Прибавьте единицу к тем элементам массива x, индексы которых указаны в массиве i. В случае, если некоторый индекс встретился в массиве i несколько раз, прибавьте к соответствующему элементу массива x число вхождений данного индекса в массив i.
import numpy as np
x = np.ones(10)
i = np.array([0, 1, 2, 3, 5, 5, 5, 8])
for k in range (len(x)):
    x[k] = x[k] + np.count_nonzero(i == k)
print(x)
```
